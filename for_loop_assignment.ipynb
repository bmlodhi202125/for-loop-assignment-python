## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
n=int(input("enter the number"))
count_even=0
count_odd=0
for i in range(1,n+1):
  if i%2==0:
    count_even+=1
  elif i%2!=0:
    count_odd+=1

print(f"even number={count_even}")
print(f"odd number={count_odd}")
print(f"total count={count_even+count_odd}")
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n=int(input("enter the number"))
for i in range(2,n):
  if n%i==0:
    print("not prime")
    break
  elif n>1:
    print("prime")
  else:
    print("not prime")

```
