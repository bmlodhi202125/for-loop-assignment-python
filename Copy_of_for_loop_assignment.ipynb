## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
n=int(input("enter the number"))
even_count=0
odd_count=0
for i in range(1,n+1):
  if i%2==0:
    even_count+=1
  elif i%2!=0:
    odd_count+=1
print(f"even count:{even_count}")
print(f"odd count:{odd_count}")
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n=int(input("enter the number"))
if n<=1:
  print("not prime")
else:

  for i in range(2,n):
    if n%i==0:
       print("not prime")
       break
  else:
    print("prime")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
n= int(input("enter the number"))
for num in range(1,n+1):
  flag=True
  for i in range(2,num):
    if num%i==0:
      flag=False
      break

  if flag:
    print(num)

```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
# write your code here
b=int(input())
for i in range(1,b):
  for j in range(1,i+1):
    print(j, end="")
  print()
```
